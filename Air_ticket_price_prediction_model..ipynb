```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_excel("C:/Users/Prasa/Desktop/project.xlsx")
```

```python
df.head()
```

```python
df.drop("Additional_Info", axis = 1, inplace = True)
```

```python
df.head()
```

```python
df.select_dtypes(include = "object")
```

```python
df["Date_of_Journey_day"] = pd.to_datetime(df["Date_of_Journey"]).dt.day
df["Date_of_Journey_month"] = pd.to_datetime(df["Date_of_Journey"]).dt.month

df["Date_of_Arrival_day"] = pd.to_datetime(df["Arrival_Time"]).dt.day
df["Date_of_Arrival_month"] = pd.to_datetime(df["Arrival_Time"]).dt.month
```

```python
df.drop(["Date_of_Journey","Arrival_Time" ], axis = 1, inplace = True)
```

```python
df.select_dtypes(include = "object")
```

```python
df.drop("Route", axis = 1, inplace = True)
```

```python
df["Total_Stops"].value_counts()
```

```python
df["Total_Stops"] = df["Total_Stops"].replace({"non-stop":0, "1 stop":1, "2 stops":2, "3 stops":3, "4 stops":4 })
```

```python
df.head()
```

```python
df.dropna(inplace = True)
```

```python
df["Total_Stops"] = df["Total_Stops"].astype("int")
```

```python
df.select_dtypes(include = "object")
```

```python
df["departure_Hours"] = pd.to_datetime(df["Dep_Time"]).dt.hour
df["departure_minutes"] = pd.to_datetime(df["Dep_Time"]).dt.minute
```

```python
df.drop("Dep_Time",axis = 1, inplace = True)
```

```python
df.select_dtypes(include = "object")
```

```python
df["Duration_hour"] = pd.to_timedelta(df["Duration"]) #converting Duration into Hours and min time formate
```

```python
df["Duration_in_min"] = pd.to_timedelta(df["Duration_hour"]).astype('timedelta64[m]').astype(int) #converting Duration into minutes.
```

```python
df.head(2)
```

```python
df.drop(["Duration_hour", "Duration"], axis = 1, inplace = True) #dropping Duration and Duration_hour as we already converted these into minutes.
```

```python
df.head(1)
```

```python
df.select_dtypes(include = "object")
```

```python
df = pd.get_dummies(df,drop_first=True )
```

```python
df.head()
```

```python
#spliting the data into x,y

x = df.drop("Price", axis = 1)
y = df["Price"]
```

```python
#Spliting the data into train and test

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.2, random_state = 34)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
#Checking out corelation Btw Features and target
plt.figure(figsize=(18,18))
sns.heatmap(df.corr(), annot = True)
```

```python
# Important feature using ExtraTreesRegressor

from sklearn.ensemble import ExtraTreesRegressor
selection = ExtraTreesRegressor()
selection.fit(x, y)
```

```python
print(selection.feature_importances_)
```

```python
#plot graph of feature importances for better visualization

plt.figure(figsize = (12,8))
feat_importances = pd.Series(selection.feature_importances_, index=x.columns)
feat_importances.nlargest(10).plot(kind='barh')
plt.show()
```

```python
#building a model using Random forrest
```

```python
from sklearn.ensemble import RandomForestRegressor
reg_rf = RandomForestRegressor()
reg_rf.fit(x_train, y_train)
```

```python
reg_rf.score(x_train, y_train) --- #Train_Score
```

```python

reg_rf.score(x_test,y_test) -- #Test Score
```

```python
y_pred = reg_rf.predict(x_test)
```

```python
sns.distplot(y_test-y_pred)
plt.show()
```

```python

plt.scatter(y_test, y_pred, alpha = 0.5)
plt.xlabel("y_test")
plt.ylabel("y_pred")
plt.show()
```

```python
from sklearn import metrics

metrics.r2_score(y_test, y_pred)
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, y_pred))
print('MSE:', metrics.mean_squared_error(y_test, y_pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
